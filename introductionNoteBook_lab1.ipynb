<h3>Agregando encabezados (Headers)</h3>

Revisa nuestro dataset. `Pandas` automáticamente prepara los encabezados con un integer comenzando desde el 0.

Para una mejor descripción de nuestra información de nuestra información, podemos introducir encabezados. esta información se encuentra disponible en:  <a href="https://archive.ics.uci.edu/ml/datasets/Automobile?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDA0101ENSkillsNetwork20235326-2021-01-01">https://archive.ics.uci.edu/ml/datasets/Automobile</a>.

Por lo tanto, debemos agregar nuestros encabezados de forma manual.

**Primero**, creamos una lista "headers" que incluye todos los nombres de las columnas en orden.
Luego, usamos `dataframe.columns = headers` para reemplazar los encabezados con la lista que creamos.


````python
# Creamos la lista headers
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
````

Reemplazamos los encabezados y volvemos a revisar nuestro dataframe.

````python
df.columns = headers
df.head(10)
````

Necesitamos reemplazar el símbolo "?" con NaN asi que usamos el método dropna() para remover los valores faltantes.

````python
# Reemplazamos los valore "?" por NaN
df1 = df.replace("?",np.NaN)

# Revisamos nuestro dataframe para ver que los valore "?" hayan sido cambiados por NaN.
df1.head()
````

Podemos votar los valores perdidos a lo largo de la columna "price" de la siguiente manera:

````python
df = df1.dropna(subset=["price"], axis=0)

# Revisamos las primeras 20 filas para asegurarnos de que no existen valores NaN en la columna "price"
df.head(20)
````

Ahora, hemos leído exitosamente un dataset en bruto y agregado correctamente los headers al dataframe.

## Describe

Si quisiéramos tener un sumario estadístico de cada columna e.j. count, valor medio de la columna, desviación estándar de la columna , etc., usamos el método describe:

**dataframe.describe()**

Este método nos proporciona un sumario estadístico variado, excluyendo los valores **NaN** (Not a Number).

````python
df.describe()
````

Esto muestra un sumario estadístico de todas las columnas de tipo numéricos 

Por ejemplo "symboling" posee 205 conteos, el valor promedio de esta columna es de 0.83, la desviación estándar es de 1.25, el valor mínimo es de -2the mean value of this column is 0.83, the standard deviation is 1.25, the minimum value is -2, El percentil 25 es 0, el percentil 50 es 1, el percentil 75 es 2 y el valor máximo es 3.

Sin embargo, que pasa si quisiéramos revisar todas las columnas incluidas aquellas de tipo object?

Puedes agregar el argumento `include = "all"` dentro del bracket. Vamos a intentarlo nuevamente.

````python
# Describimos todas las columnas en "df"

df.describe(include="all")
````

Ahora nos entrega el sumario estadístico de todas las columnas, incluyendo los atributos tipo-object

Ahora podemos ver cuantos valores únicos hay, cual de ellos es el valor mas alto y la frecuencia del valor mas alto en las columnas tipo-object

Algunos valores en la tabla de arriba son mostrados como "NaN". Esto es por que esos números no están disponibles independiente de la columna en particular.
